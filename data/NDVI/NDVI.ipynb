#### Normalized Difference Vegetation Index

```python
import pandas as pd
```

```python
# load the file
modis = pd.read_csv("US_State_Monthly_NDVI_2015_2025.csv")

# keep only relevant columns
modis_clean = modis[['NAME', 'STUSPS', 'year', 'month', 'mean']].copy()

# drop missing or invalid rows
modis_clean = modis_clean.dropna(subset=['mean', 'year', 'month'])

# convert NDVI scale (MODIS NDVI is scaled by 0.0001)
modis_clean['NDVI'] = modis_clean['mean'] * 0.0001

# clean column names
modis_clean = modis_clean.rename(columns={
    'NAME': 'State',
    'STUSPS': 'StateCode',
    'year': 'Year',
    'month': 'Month'
})[['State', 'StateCode', 'Year', 'Month', 'NDVI']]

# sort by state and date
modis_clean = modis_clean.sort_values(['StateCode', 'Year', 'Month'])

# clean and save data
modis_clean.to_csv("NDVI_state_monthly_clean_2018_2025.csv", index=False)

modis_clean
```
